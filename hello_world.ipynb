```python
!pip install pycuda
import pycuda.autoinit
import pycuda.driver as cuda
from pycuda.compiler import SourceModule
```

Lets define our kernel.

```python
modd = SourceModule("""__global__ void print_from_gpu(){
  if (threadIdx.x % 2 == 0)
    {
    printf("Hello world from CUDA! Thread: %d, Block: %d", threadIdx.x, blockIdx.x);
    }
}
""")
```

Lets call the CUDA kernel.

```python
kernel = modd.get_function("print_from_gpu")
kernel(block=(5,1,1), grid=(1,1,1))
```
